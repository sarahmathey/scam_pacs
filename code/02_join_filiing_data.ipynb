# Join all filing data together (SB, SE)

```python
import pandas as pd
from pathlib import Path

# Define the base directory containing subdirectories
base_dir = Path("/Users/sarahmathey/Desktop/DSII/final_project/data/raw/output")


# Lists to store data for SB and SE datasets
sb_data = []
se_data = []
# Run Through all subdirectories
for sub_dir in base_dir.iterdir():
    # If it's a directory
    if sub_dir.is_dir():  
        # Extract directory name as filing_id to add to dataset
        filing_id = sub_dir.name  
        
        # Run through CSV files in the subdirectory
        for csv_file in sub_dir.glob("*.csv"):
            # if the file name starts with SB, 
            if csv_file.name.startswith("SB"):
                # create a df with csv
                df = pd.read_csv(csv_file)
                # append filing_id as column  
                df["filing_id"] = filing_id  # Add filing_id column
                # add to sb_data list
                sb_data.append(df)
            # If the file name starts with SE,
            elif csv_file.name.startswith("SE"):
                #create a df with the csv
                df = pd.read_csv(csv_file)
                #append the filing id as a columsn 
                df["filing_id"] = filing_id  # Add filing_id column
                #add new df to se_data list 
                se_data.append(df)

# Combine all dataframes from lists into final datasets & export as CSV
if sb_data:
    sb_dataset = pd.concat(sb_data, ignore_index=True)
    sb_dataset.to_csv("SB_dataset.csv", index=False)

if se_data:
    se_dataset = pd.concat(se_data, ignore_index=True)
    se_dataset.to_csv("SE_dataset.csv", index=False)

print("Processing complete. SB and SE datasets saved.")
```
